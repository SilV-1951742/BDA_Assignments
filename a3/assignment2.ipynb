# Prepare parallel data

Hier wordt de relevante data van de parallelle runs uitgelezen en klaargemaakt voor gebruik.

````python
content = None
tmp_times = []
amount_threads_used = np.array([1, 2, 4, 8, 16, 32])
amount_runs = 6
times_per_run = 6

with open('a_all.pbs.o50916276') as f:
    content = f.readlines()

for line in content:
    if 'sequential' in line:
        words = line.split(',')
        time = words[len(words) - 1]
        tmp_times.append(float(time))

times = np.array(tmp_times)
#print(times)
````

````python
runs = np.split(times, times_per_run)
#print(runs)
````

````python
frame = np.empty(shape=(3, amount_runs))

frame[0,:] = amount_threads_used
frame[1,:] = np.average(runs, axis=1)
frame[2,:] = np.std(runs, axis=1)

#print(frame)
````

````python
frame_tuples = []
for i in range(0, amount_runs):
    for elem in runs[i]:
        frame_tuples.append((amount_threads_used[i], elem))
#print(frame_tuples)
````

# Prepare serial data

Hier wordt de relevante data van de seriële runs uitgelezen en klaargemaakt voor gebruik.

````python
content_s = None
tmp_times_s = []
amount_threads_used_s = np.array([1])
amount_runs_s = 1
times_per_run_s = 5

with open('a_all.pbs.o50917985') as f:
    content_s = f.readlines()

for line in content_s:
    if 'sequential' in line:
        words = line.split(',')
        time = words[len(words) - 1]
        tmp_times_s.append(float(time))

times_s = np.array(tmp_times_s)
#print(times_s)
````

````python
frame_s = np.empty(shape=(3, amount_runs_s))

frame_s[0,:] = amount_threads_used_s
frame_s[1,:] = np.average(times_s)
frame_s[2,:] = np.std(times_s)

#print(frame_s)
````

````python
expected_time = np.zeros(shape=(6, amount_runs))
expected_time.fill(frame_s[1,0])
serial_full = np.zeros(shape=(6, amount_runs))
idx = 0
for array in expected_time:
    serial_full[idx] = array
    expected_time[idx] = array / amount_threads_used[idx]
    idx += 1
#print(expected_time)
#print(serial_full)
````

# Vergelijking seriële implementie vs OpenMP single threaded programma

Bij de normale serial implementatie ligt de uitvoertijd op 6.93 seconden met een standaarddeviatie van 0.067 seconden terwijl de OpenMP single threaded implementatie op 7.76 seconden ligt met een standaarddeviatie van 1.11 seconden.

De normale serial implementatie is dus beduidend sneller dan de OpenMP implementatie door de extra overhead. De hogere standaarddeviatie komt volgens ons ook van het aanmaken van de threadpool en de availability van threads.

# Speedup en efficiency

Zoals eerder gezegd zijn er voor elk datapunt zes runs gedaan, dus zes runs op 1, 2, 4, 8, 16 en 32 cores.

Er is te zien dat de efficiency snel afvalt af en dat de speedup niet echt snel stijgt. Dit valt volgens ons te wijten aan het feit dat de code sterk is geoptimaliseerd voor seriële performance. We hebben geprobeerd hieraan te sleutelen maar veel verandering kwam er niet en we wilden de code wel in dezelfde geest houden als de originele.

Wel blijft de speedup en efficiency doorheen 

````python
sup_arr = serial_full / runs

sup_tuples = []
for i in range(0, amount_runs):
    for j in range(0, times_per_run):
        sup_tuples.append((amount_threads_used[i], sup_arr[i][j], amount_threads_used[i]))
        
# print(sup_tuples)
````

````python
df_sup = pd.DataFrame(sup_tuples, columns=["Threads", "True speedup", "Expected speedup"])
df_sup.Threads = df_sup.Threads.astype(int)

df_sup.set_index("Threads", inplace=True)
# print(df_sup)
````

````python
plt.title("True vs expected speedup")
plt.ylabel("Speedup")
plt.xlabel("Amount of threads")

sns.lineplot(data=df_sup)
#plot.set_xscale('log')
````

````python
eff_tuples = []
for elem in sup_tuples:
    eff_tuples.append((elem[0], elem[1] / elem[0], elem[2] / elem[0]))
    
# print(eff_tuples)
````

````python
df_eff = pd.DataFrame(eff_tuples, columns=["Threads", "True efficiency", "Expected efficiency"])
df_eff.Threads = df_eff.Threads.astype(int)

df_eff.set_index("Threads", inplace=True)
# print(df_eff)
````

````python
plt.title("True vs expected efficiency")
plt.ylabel("Efficiency")
plt.xlabel("Amount of threads")

sns.lineplot(data=df_eff)
````

# Serial fraction

Als we de formule voor de speedup omvormen krijgen we:
$$s = \frac{N}{(N-1) \cdot SpeedUp} - \frac{1}{N - 1}$$

Hieronder berekenen we de gemiddelde speedup voor acht threads, deze is 2,54.
Als we de bovenstaande formule invullen krijgen we:
$$\begin{eqnarray} 
s &= \frac{8}{(8-1) \cdot 2.54} - \frac{1}{8 - 1} \\ 
s &= 0.31
\end{eqnarray} $$

Dus 30% van het programma zou serieel zijn. Als we onderstaande empirisch gevonden resultaten bekijken lijkt dit niet te kloppen.

Nu we seriële fractie hebben kunnen we de speedup van 16 threads berekenen:

$$\begin{eqnarray}
SpeedUp &= \frac{1}{s + \frac{p}{N}}\\
SpeedUp &= \frac{1}{0,31 + \frac{0,69}{16}} \\
SpeedUp &= 2,83
\end{eqnarray}$$

Dus bij een verhoging van 8 naar 16 threads zouden we een speedup van 2,81 verwachten, dit is incorrect en komt omdat de seriële fractie te hoog ligt bij voorgaande berekening.

Als we de uitvoertijd van een thread nemen, deze is 7,76 seconden, en delen deze door de speedup van 2,81 krijgen we: 2,76 seconden wat te hoog is. In praktijk krijgen we een gemiddelde uitvoertijd van 1,29 seconden.

````python
avg_e_threads_sup = 0
amount_e_ts = 0
for elem in sup_tuples:
    if elem[0] == 8:
        amount_e_ts += 1
        avg_e_threads_sup += elem[1]
        
avg_e_threads_sup = avg_e_threads_sup / amount_e_ts

print(avg_e_threads_sup)
````

````python
serial_frac = (8 / ((8-1) * 2.54)) - (1 / (8 - 1))
print(serial_frac)
````

# Verwachtingen meerdere cores

Als we kijken naar de resultaten van de serial fraction verwachten we dat de uitvoertijd bij deze berekening ca. 33 - 45 ms plus de tijd van de berekeningen op een datapunt. Dit is uiteraard niet waar door de overhead van de threads.

Bij meer en meer cores zijn er uiteraard ook diminishing returns, zoals al zichtbaar is bij het schalen van 16 naar 32 cores, dus de code zal dan meer geoptimaliseerd moeten worden voor parallel te runnen.
